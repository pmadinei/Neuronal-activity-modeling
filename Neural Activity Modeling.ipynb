This next function should look very familiar: it's the same as in Homework 1, except for one minor upgrade (it can now deal with empty spike trains):

```python
def plot_spike_train(spike_times, t_end=None, ax=None, labels=True):
    """Plot a spike train

    Parameters
    ----------
    spike_times : array
        Array of spike times (in milliseconds)
    t_end : float
        Spike train duration (in milliseconds)
    ax : Matplotlib axis
        If None, will plot on current axes
    """
    # No axes given, use the currently active ones:
    if ax is None:
        ax = plt.gca()
    if len(spike_times) > 0:
        # If no duration is given, assume it is the last spike time:
        if t_end is None:
            t_end = spike_times[-1]
        # For each spike, plot a vertical line:
        for t in spike_times:
            ax.plot([t, t], [0, 1], 'k', linewidth=2)
        ax.set_xlim(0, t_end)
    if labels:
        ax.set_xlabel('time (ms)')
    ax.set_yticks([])
```

You need not worry about the next function. We will use it to visualize network activity, but a lot of it is just about how to deal with Matplotlib subplots.

```python
from matplotlib.patches import ConnectionPatch

def plot_network_activity(g_pre, r_post, wts, fig=None, linestyle='solid'):
    """Plot network activity

    Parameters
    ----------
    g_pre : TxM np.ndarray
        Conductance of all M presynaptic neurons as it evolves over time T
    r_post : TxN np.ndarray
        Firing rate of all N postsynaptic neurons as it evolves over time T
    wts : NxM np.ndarray
        Synaptic weights from M presynaptic to N postsynaptic neurons
    fig : Matplotlib figure or None
        If None, a new figure will be created. If given, that same figure will
        be used (useful to plot multiple trials in the same panel)
    linestyle : Matplotlib linestyle
        The line style to use, such as 'solid' or 'dashed'
    """
    if fig is None:
        # When no figure is given, create a new one:
        fig, axes = plt.subplots(nrows=3, ncols=3, sharex=True, figsize=(18, 6))
    else:
        # When a figure is given, find its axes:
        fig = plt.gcf()
        axes = np.array(fig.axes).reshape((3, 3))

    # Use different (but consistent) colors for the three different input /
    # output neurons:
    color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

    # Plot all presynaptic neurons:
    for n in range(num_pre):
        # To identify spikes, we take the gradient of `g_pre` and find the
        # array elements where the gradient is 1:
        plot_spike_train(np.where(np.isclose(np.diff(g_pre[:, n]), 1))[0] * dt, 
                         ax=axes[n, 0], t_end=t_dur, labels=False)
        axes[n, 1].plot(time, g_pre[:, n], linewidth=2, linestyle=linestyle,
                        color=color_cycle[n])
        axes[n, 1].set_ylim(-1, 5)
        axes[n, 1].set_xlabel('time (ms)')

    # Plot all postsynaptic neurons:
    for n in range(num_post):
        axes[n, 2].plot(time, r_post[:, n], linewidth=2, linestyle=linestyle,
                        color=color_cycle[n])
        axes[n, 1].set_xlabel('time (ms)')
        axes[n, 0].set_ylabel(f'Neuron {n}')
        axes[n, 2].set_ylim(r_post.min() - 0.1 * np.abs(r_post.max()), 
                            1.1 * r_post.max())
    axes[0, 0].set_title('Poisson spikes')
    axes[0, 1].set_title('pre: $g_s$')
    axes[0, 2].set_title('post: $r_\mathrm{out}$')
    
    # All this stuff is just to show arrows between subpanels to visualize the
    # connectivity matrix `wts`:
    for row in range(3):
        con = ConnectionPatch((1.02, 0.5), (-0.05, 0.5), axesA=axes[row, 0], 
                              axesB=axes[row, 1], coordsA='axes fraction', 
                              coordsB='axes fraction', arrowstyle='->')
        axes[row, 1].add_artist(con)
    for post in range(wts.shape[0]):
        for pre in range(wts.shape[1]):
            if not np.isclose(wts[post, pre], 0, atol=1e-5):
                if wts[post, pre] > 0:
                    arrowstyle = '-|>'
                    linestyle = 'solid'
                else:
                    arrowstyle = '-'
                    linestyle = 'dashed'
                w = wts[wts > 1e-5]
                if not np.isclose(np.max(w) - np.min(w), 0):
                    linewidth = (wts[post, pre] - np.min(w)) / (np.max(w) - np.min(w))
                else:
                    linewidth = 0.2
                con = ConnectionPatch((1.02, 0.5), (-0.1, 0.5), 
                                      axesA=axes[pre, 1], axesB=axes[post, 2], 
                                      coordsA='axes fraction',
                                      coordsB='axes fraction',
                                      arrowstyle=arrowstyle, color='k', 
                                      linewidth=2.5 * linewidth + 0.5,
                                      linestyle=linestyle)
                axes[post, 2].add_artist(con)
```

# 1. Firing-Rate Neurons 

In this first section, we will create a simple network of firing-rate neurons.

We will consider a trial duration of 2 seconds (`t_dur`), and we will output network activity in steps of `dt` milliseconds:

```python
dt = .1
t_dur = 2000
time = np.arange(0, t_dur, dt)
```

Our network will consist of 3 presynaptic neurons (firing Poisson spikes at a given rate, `r_pre`) with an exponential synaptic kernel (exponential decay constant, `tau_pre`) as well as 3 postsynaptic neurons with their own firing-rate dynamics (`tau_post`):

```python
num_pre = 3
r_pre = np.array([1, 2, 10])
tau_pre = 50
```

## 1.2 TODO Simulating Network Activity 

The next step is to simulate the network activity. We will need to create a function that takes the Poisson firing rates (`r_pre`), the connectivity matrix (`wts`), and the taus as input (`tau_pre`, `tau_post`). The function should then return the synaptic conductance (`g_pre`) and postsynaptic firing rate (`r_post`), evaluated at each time step in `dt` increments.

This brings us back to Lecture 8, where we had (in this HW: Poisson) spikes as input and want to filter them with a synaptic kernel:

$ g_\mathrm{pre} = \int_{-\infty}^{t} K_s(t-\tau) \rho(\tau) d\tau $

where $\rho$ is the neural response function and $K_s(t) = exp(-t / \tau_\mathrm{pre})$.

To be clear, this can be implemented with a differential equation:

$ \frac{dg_\mathrm{pre}}{dt} = g_\mathrm{pre} exp(-dt / \tau_\mathrm{pre}) + \delta(t)$

where $\delta(t)$ is 1 if a spike happened at time $t$, and 0 otherwise. 

To be *very* clear, in the below code $g_\mathrm{pre}$ should therefore do the following at each time step:
- decay exponentially with time constant `tau_pre`
- in the presence of a spike (generated with the Poisson process), increase $g_\mathrm{pre}$ by +1

The postsynaptic firing rate $r_\mathrm{post}$ should behave similarly:
- decay exponentially with time constant `tau_post`
- increase by the dot product of synaptic weights times $g_\mathrm{pre}$

This means that we are not implementing an activation function $F$ (or more precisely: assume that $F(x)=x$).

```python
def run_network(r_pre, wts, tau_pre, tau_post):
    """Firing-rate network

    Parameters
    ----------
    r_pre : Mx1 np.ndarray
        Each presynaptic neuron should have a firing rate, according to which
        Poisson spikes will be generated
    wts : NxM np.ndarray
        Connectivity matrix
    tau_pre : float
        Exponential decay constant in ms (synaptic conductance)
    tau_post : float
        Exponential decay constant in ms (postsynaptic firing rate)

    Returns
    -------
    g_pre : TxM np.ndarray
        Synaptic conductance for all M presynaptic neurons, evaluated at all
        time points T
    r_post : TxN np.ndarray
        Firing rate for all N postsynaptic neurons, evaluated at all time
        points T
    """
    # Make sure to use the same random seed every time:
    np.random.seed(1234)

    # Initialize with 0:
    g_pre = np.zeros((len(time),num_pre))
    r_post = np.zeros((len(time),num_post))

    for t in range(1, len(time)):
        # Decay synaptic conductances for all presynaptic neurons. Here the
        # value at time t-1 should be used to inform the value at time t.
        # Also, make sure the values don't go below 0
        g_pre[t, :] = np.maximum(0, np.exp(-dt/tau_pre) * g_pre[t-1, :])
        r_post[t, :] = np.maximum(0, np.exp(-dt/tau_post) * r_post[t-1, :])

        # This is the easiest way to generate Poisson spikes:
        has_spiked = np.random.rand(num_pre) < r_pre * dt / 1000.

        # We need to update the synaptic conductance for all presynaptic neurons
        # that have spiked:
        g_pre[t, has_spiked] = g_pre[t-1, has_spiked] + 1

        # Finally, we need to update the postsynaptic firing rate. Make sure the
        # firing rate does not go below 0
        r_post[t] = np.maximum(0, r_post[t] + dt * wts@g_pre[t,:])
    return g_pre, r_post
```

## 2.1 TODO STDP Curve (2 pts)

The first step is to implement the STDP curve.
We will use an LTD window that is slightly larger than the LTP window, to discourage runaway potentiation:

```python
# LTP:
A_plus = 0.05
tau_ltp = 20

# LTD:
A_minus = -A_plus
tau_ltd = 40
```

We will then create an array of time differences and evaluate the STDP curve on each element:

```python
time_diff = np.linspace(-5 * tau_ltd, 5 * tau_ltp)
```

We will do this in a function. The function should accept the array of time differences and all STDP parameters, and it should return the STDP-dictated weight change for each time difference value:

```python
def stdp(time_diff, A_plus, A_minus, tau_ltp, tau_ltd):
    """STDP weight change

    Parameters
    ----------
    time_diff : np.ndarray
        Array of time differences
    A_plus, A_minus : float
        excitatory, inhibitiory peaks of the STDP curve
    tau_stdp : float
        exponential decay constant
    """
    stdp = np.zeros_like(time_diff)
    stdp[time_diff > 0] = A_plus * np.exp(-(time_diff[time_diff > 0]) / tau_ltp)
    stdp[time_diff < 0] = A_minus * np.exp((time_diff[time_diff < 0]) / tau_ltd)
    return stdp
```

```python
plt.plot(time_diff, stdp(time_diff, A_plus, A_minus, tau_ltp, tau_ltd), linewidth=2)
plt.grid('major')
plt.xlabel('$\Delta t$')
plt.ylabel('$\Delta w$')
```

## 2.3 TODO Simulating Network Activity (6 pts)

STDP doesn't work with firing-rate neurons, so we need to implement a spiking neuron. Let's bring back our good old friend, the Izhikevich neuron.

For now, don't worry about STDP. The following function is building on the `run_network` function we wrote above - except now it should implement postsynaptic neurons as Izhikevich neurons rather than firing-rate neurons.

```python
def run_spiking_network(r_pre, wts, tau_pre, a=.02, b=.2, c=-65, d=8):
    """Spiking neural network

    Parameters
    ----------
    r_pre : Mx1 np.ndarray
        Each presynaptic neuron should have a firing rate, according to which
        Poisson spikes will be generated
    wts : NxM np.ndarray
        Connectivity matrix
    tau_pre : float
        Exponential decay constant in ms (synaptic conductance)
    a, b, c, d : float
        Izhikevich parameters

    Returns
    -------
    g_pre : TxM np.ndarray
        Synaptic conductance for all M presynaptic neurons, evaluated at all
        time points T
    v_post : TxN np.ndarray
        Membrane potential for all N postsynaptic neurons, evaluated at all time
        points T
    """
    # Make sure to use the same random seed every time:
    np.random.seed(1234)

    # Initialize with 0:
    g_pre = np.zeros((len(time), num_pre), dtype=float)
    v_post = np.zeros((len(time), num_post), dtype=float)
    u_post = np.zeros((len(time), num_post), dtype=float)

    # Initialize regular-spiking Izhikevich neuron:
    v0 = -70
    v_post[0, :] = v0
    u_post[0, :] = b * v0

    for t in range(1, len(time)):
        # Decay synaptic conductances for all presynaptic neurons. Here the
        # value at time t-1 should be used to inform the value at time t.
        # Also, make sure the values don't go below 0
        g_pre[t, :] = np.maximum(0, np.exp(-dt/tau_pre) * g_pre[t-1,:])

        # This is the easiest way to generate Poisson spikes:
        has_pre_spiked = np.random.rand(num_pre) < r_pre * dt / 1000.

        # We need to update the synaptic conductance for all presynaptic neurons
        # that have spiked:
        g_pre[t, has_pre_spiked] = g_pre[t-1, has_pre_spiked] + 1

        # The synaptic conductance just has values between 0 and 1, so we need
        # to scale it with the synaptic weights to get the current:
        I = wts @ g_pre[t,:]

        # Finally, we need to update the postsynaptic neuron. Here we will use
        # a regular-spiking Izhikevich neuron:
        v = v_post[t - 1, :]
        u = u_post[t - 1, :]
        v_post[t, :] = v + dt * (0.04 * (v**2) + 5 * v + 140 - u + I)
        u_post[t, :] = u + dt * (a * (b * v - u))

        # If one or more of the neurons spiked, we need to update v and u:
        has_post_spiked = v_post[t, :] > 30
        v_post[t - 1, has_post_spiked] = 30
        v_post[t, has_post_spiked] = c
        u_post[t, has_post_spiked] = u_post[t, has_post_spiked] + d

    return g_pre, v_post
```

```python
g_pre, v_post = run_spiking_network(r_pre, wts, tau_pre)
```

## 2.3 TODO Simulating STDP (6 pts)

The final step is to implement STDP. For this we will update our network function once again, but this time we will add in the LTP/LTD updates.

```python
def run_stdp_network(r_pre, wts, a=.02, b=.2, c=-65, d=8):
    """Spiking neural network with STDP

    Parameters
    ----------
    r_pre : Mx1 np.ndarray
        Each presynaptic neuron should have a firing rate, according to which
        Poisson spikes will be generated
    wts : NxM np.ndarray
        Connectivity matrix
    a, b, c, d : float
        Izhikevich parameters

    Returns
    -------
    g_pre : TxM np.ndarray
        Synaptic conductance for all M presynaptic neurons, evaluated at all
        time points T
    v_post : TxN np.ndarray
        Membrane potential for all N postsynaptic neurons, evaluated at all time
        points T
    """
    # Make sure to use the same random seed every time:
    np.random.seed(123)

    # Initialize with 0:
    g_pre = np.zeros((len(time), num_pre), dtype=float)
    v_post = np.zeros((len(time), num_post), dtype=float)
    u_post = np.zeros((len(time), num_post), dtype=float)

    # We will also need to evolve the weights over time. We initialize (t=0)
    # the weights to `wts` given by the connectivity matrix:
    wts_stdp = np.zeros((len(time), num_post, num_pre), dtype=float)
    wts_stdp[0, :, :] = wts

    # Initialize regular-spiking Izhikevich neuron:
    v0 = -70
    v_post[0, :] = v0
    u_post[0, :] = b * v0

    # To run STDP correctly, we need to keep track of the last spike time for
    # each neuron. We will initialize it with a large negative value, so that
    # it for sure falls outside the LTP/LTD window even if we get a spike at
    # t=0:
    last_spike_pre = np.zeros(num_pre) - 1000
    last_spike_post = np.zeros(num_post) - 1000

    for t in range(1, len(time)):
        # We need a new variable that keeps track of all the weight changes
        # for this time step:
        dwts = np.zeros((num_post, num_pre))

        # Decay synaptic conductances for all presynaptic neurons. Here the
        # value at time t-1 should be used to inform the value at time t.
        # Also, make sure the values don't go below 0
        g_pre[t, :] =  np.maximum(0, np.exp(-dt / tau_pre) * g_pre[t-1, :])

        # This is the easiest way to generate Poisson spikes:
        has_pre_spiked = np.random.rand(num_pre) < r_pre * dt / 1000.
        if np.any(has_pre_spiked):
            # We need to update the synaptic conductance for all presynaptic
            # neurons that have spiked:
            g_pre[t, has_pre_spiked] = g_pre[t-1, has_pre_spiked] + 1

            # Possible LTD: We need to call the STDP function with the
            # appropriate \Delta t, which here is called `time_diff`. 
            # For this we need to update the last spike time for all presynaptic
            # neurons that actually spiked just now:
            last_spike_pre[has_pre_spiked] = time[t]
            # Of course, it might so happen that every presynaptic neuron spiked
            # just now, so we need to calculate the LTP contribution on every 
            # synapse and compare it to the last postsynaptic spike.
            # You can do this with a for loop, or more elegantly with:
            time_diff = last_spike_post[:, np.newaxis] - last_spike_pre
            # The `np.newaxis` thing will make it so that the postsynaptic spike
            # time will be compared to every presynaptic spike time. The
            # resulting `time_diff` is a matrix that has the same shape as 
            # `dwts`. So all you need to do is call the STDP function here with
            # the appropriate values:
            dwts += stdp(time_diff, A_plus, A_minus, tau_ltp, tau_ltd)

        I = wts @ g_pre[t, :]

        # Finally, we need to update the postsynaptic neuron. Here we will use
        # a regular-spiking Izhikevich neuron:
        v = v_post[t - 1, :]
        u = u_post[t - 1, :]
        v_post[t, :] = v + dt * (0.04 * (v**2) + 5 * v + 140 - u + I)
        u_post[t, :] = u + dt * (a * (b * v - u))

        # If one or more of the neurons spiked, we need to update v and u:
        has_post_spiked = v_post[t, :] > 30
        if np.any(has_post_spiked):
            v_post[t - 1, has_post_spiked] = 30
            v_post[t, has_post_spiked] = c
            u_post[t, has_post_spiked] = u_post[t, has_post_spiked] + d

            # Possible LTP:
            last_spike_post[has_post_spiked] = time[t]
            time_diff = last_spike_post[:, np.newaxis] - last_spike_pre
            dwts += stdp(time_diff, A_plus, A_minus, tau_ltp, tau_ltd)

        wts_stdp[t, :, :] = wts_stdp[t - 1, :, :] + dwts

    return g_pre, wts_stdp, v_post
```

To see how STDP reacts to spiking input of different frequencies, let's update the firing rate of the presynaptic neurons slightly. Then we run the network:

```python
r_pre_stdp = np.array([0, 2, 30])
g_pre, wts_stdp, v_post = run_stdp_network(r_pre_stdp, wts)
```
